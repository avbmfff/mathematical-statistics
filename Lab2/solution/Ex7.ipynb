```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
counts = []
props = []
errorprobs = []
theta = 0.5
n_volume = [50, 100, 500, 1000, 2500] 
thetas = np.empty([5, 500], dtype=np.float64)
```

```python
for i, n in enumerate(n_volume):
    count = 0
    prop = 0
    errorprob = 0
    for j in range(500):
        sample = np.random.geometric(theta, n)
        y = sample ** 2
        theta2 = 1 / np.mean(sample)
        thetas[i, j] = theta2
        if np.abs(theta - theta2) > 0.01:
            count += 1
            prop = np.mean(thetas, axis = 1)  
            errorprob = prop.mean()      
    counts.append(count)
    props.append(prop)
    errorprobs.append(errorprob)
```

```python
result = pd.DataFrame(data={'n': n_volume, 'count': counts, 'error probality': errorprobs})
result
```

```python
px.bar(result, x=list(map(str, n_volume)), y='count', title='Dependence of the parameter on the sample size')
```

Вывод: при увеличении объёма выборки оценка становится точнее (число отклонений от истинного значения уменьшается). Это происходит из-за того, что при большом объёме выборки оценка описывает истинное распределние точнее.
