Дано: $N(\mu_1, σ^2_1), N(\mu_2, σ^2_1)$ объёмов соответсвенно $m$ и $n$. $τ = μ_1-\mu_2, σ^2_1=σ^2_1$ неизвестна, $\mu_1=2, \mu_2=1$.
Функция: $\sqrt{\frac{mn(m+n-2)}{m+n}}\frac{\overline{X}-\overline{Y}-τ}{\sqrt{nVar(x)+mVar(Y)}}$

### Решение

$x \sim N(\mu_1, σ_1^2)$

$y \sim N(\mu_2, \sigma_2^2)$

$\sqrt{n} \frac{\overline{X}-\mu_1}{σ_1} \sim N(0,1)$ ⇒ $\overline{X} \sim N(\mu_1, \frac{\sigma_1^2}{n})$

$\sqrt{m} \frac{\overline{Y}-\mu_2}{σ_2} \sim N(0,1)$  ⇒ $\overline{Y} \sim N(\mu_2, \frac{\sigma_1^2}{m})$

$E(\overline{X})-E(\overline{Y}) = E(X) - E(Y) = \mu_1 - \mu_2 = τ$

$D(\overline{X}-\overline{Y}) = D(\overline{X}) + D(\overline{Y}) = \frac{σ^2}{n} + \frac{\sigma^2}{m}$

$\overline{X} - \overline{Y} \sim N(τ, \frac{σ^2}{n} + \frac{\sigma^2}{m})$

$\overline{X} - \overline{Y} - τ \sim N(0, \frac{σ^2}{n} + \frac{\sigma^2}{m})$

$\frac{\overline{X}-\overline{Y} - \tau}{\sqrt{\frac{σ^2}{n} + \frac{\sigma^2}{m}}} \sim N(0, 1)$

$\text{По теореме Фишера:}$

$\frac{nD(X)}{σ^2}\space \text{и} \space \frac{mD(Y)}{σ^2} \sim χ^2(n-1), χ^2(m-1), \space \text{так как их сумма} \sim \chi^2(n+m-2)$

$\sqrt{\frac{mn(m+n-2)}{m+n}}\frac{\overline{X}-\overline{Y}-τ}{\sqrt{nVar(x)+mVar(Y)}} = \frac{(\overline{X}-\overline{Y}-τ)/\sqrt{\frac{σ^2}{n} + \frac{\sigma^2}{m}}}{\frac{nD(X)+mD(y)}{σ^2(n+m-2)}}$

$\sqrt{\frac{mn(m+n-2)}{m+n}}\frac{\overline{X}-\overline{Y}-τ}{\sqrt{nVar(x)+mVar(Y)}} \sim \space \text{распределение стьюдента с n+m-2 степенями свободы}$.

### Код

```python
import numpy as np
import statistics as stat
import scipy
```

```python
def sample_mean(size, array):
  sum = 0
  for i in range(size):
    sum += array[i]
  return sum/size
```

```python
def get_dispersion(size, array, mean):
  sum = 0
  for i in range(size):
    sum += (array[i]-mean)**2
  return sum/size
```

```python
def get_sample(size, mu, sigma):
  return mu + sigma * np.random.standard_normal(size = size)
```

Эксперимента для $m = 25, n= 25$

```python
size_x = 25
size_y = size_x
mu_x = 2
mu_y = 1
sigma_x = 1
sigma_y = sigma_x
t = mu_x - mu_y
```

```python
count = 0
for i in range(1000):
  x = get_sample(size_x, mu_x, sigma_x)
  y = get_sample(size_y, mu_y, sigma_y)

  x_mean = sample_mean(size_x, x)
  y_mean = sample_mean(size_y, y)
  
  d_e = get_dispersion(size_x, x, x_mean)
  s = ((size_x/(size_x-1))*d_e)**(0.5)
  delta = (scipy.stats.t.ppf((1+0.95)/2, size_x-1) * s) / size_x**(0.5)


  left = (x_mean - y_mean) - delta
  right = (x_mean - y_mean) + delta

  if t <= right and t >= left:
    count += 1

print(count)
```
