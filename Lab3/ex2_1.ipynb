Дано: $Exp(λ)$, оцениваемый параметр - медиана, $\lambda = 1$

$\hat{\mu} = \overline{X} ≃ N(\mu, \frac{\sigma^2}{n}) ⇔$ [центрирование] $\overline{X}-\mu ≃ N(0, \frac{\hat{\sigma^2}}{n}) ⇔$  [нормирование] $\frac{\overline{X}-\mu}{\sqrt{\frac{\hat{\sigma^2}}{n}}} ≃ N(0,1)$

$P(-Z_{1-\frac{α}{2}}\leq \frac{\overline{X}-\mu}{\sqrt{\frac{\hat{\sigma^2}}{n}}} \leq Z_{1-\frac{α}{2}}) = 1 - α$

В экспоненциальном распределении $μ = \frac{1}{\lambda}$, а $med = \frac{Ln(2)}{\lambda}$. Приведем к нужному виду доверительный интервал.

После некоторых преобразований, он принимает вид: $P(\overline{X}-Z_{1-\frac{α}{2}}⋅\frac{\hat{\sigma}}{\sqrt{n}} \leq \mu \leq \overline{X}+Z_{1-\frac{α}{2}}⋅\frac{\hat{\sigma}}{\sqrt{n}})=1-α$

Подставим значения и умножим на $ln(2)$: $P(\overline{X}-Z_{1-\frac{α}{2}}⋅\frac{\hat{\sigma}}{\sqrt{n}}\cdot ln(2) \leq \frac{ln(2)}{\lambda} \leq \overline{X}+Z_{1-\frac{α}{2}}⋅\frac{\hat{\sigma}}{\sqrt{n}} \cdot ln(2))=1-α$

```python
import math
import numpy as np
import statistics as stat
from scipy. stats import expon
```

```python
def sample_mean(size, array):
  sum = 0
  for i in range(size):
    sum += array[i]
  return sum/size
```

```python
def get_sample(size, lamda):
  return expon.rvs(scale = lamda, size = size )
```

```python
lamda = 1
z = 1.96
```

Эксперимент для $size = 25$

```python
count = 0
size = 25
med = 1/lamda
for i in range(1000):
  x = get_sample(size, lamda)

  x_mean = sample_mean(size, x)

  left = x_mean - z * (1/(size**(0.5)*lamda**4))
  right = x_mean + z * (1/(size**(0.5)*lamda**4))
  if med <= right and med >= left:
    count += 1

print(count)
```
