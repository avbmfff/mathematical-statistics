$Вариант: U[-2Θ; 3Θ], Θ ∈ (0;5), Θ=3$ - равномерное непрерывное распределение.


Из условия знаем: $MX=\frac{a+b}{2}=1.5$, $DX=\frac{(b-a)^2}{12}=18.75$, где $a=-2Θ$, $b=3Θ$.

Также нам известно, что $D\overline{X}=\frac{DX}{n}=\frac{σ^2}{n}$, $\overline{X}=\frac{∑{X_i}}{n}$


Запишем условие задачи: $P[|\overline{X}-MX|≤Ε]≥1-δ$, и применим теорему Ляпунова.

$P[|\overline{X}-MX|≤Ε]≥1-δ  \Rightarrow P[-Ε≤\overline{X}-MX≤Ε]≥1-δ \Rightarrow P[-Ε+MX≤\overline{X}≤Ε+MX]≥1-δ ⇒ P[-Ε+MX≤Φ(t_2)-Φ(t_1)≤Ε+MX]≥1-δ  $

Отсюда: $t_1=\frac{-Ε}{σ_\overline{X}}$, $t_2=\frac{Ε}{σ_\overline{X}}$

Исходя из $Φ(-X)=1-Φ(X)$, получим $Φ(t)≥1-\frac{δ}{2}$, где $t=\frac{E}{σ}.$ 

Перейдём к $α-$квантиль: $\frac{E}{σ}=q(1-\frac{δ}{2})$ или $n=\frac{DX(q(1-\frac{δ}{2})^2)}{E^2}$

```python
import numpy as np 
import scipy.stats as sps
from scipy.stats import norm
import matplotlib as mplt
import matplotlib.pyplot as plt
```

```python
DX = 18.75
E = 0.01
delta = 0.05
q = 1 - delta/2
quant = norm.ppf(q)

n = (quant**2 * DX)/ E**2
print(int(n))
```

Проверим выборочное среднее по условию: $|\overline{X}-EX|>\varepsilon$ и получим конечное количество.

```python
global_size = 500
count = 0
MX = 1.5
n = 720273
for i in range(global_size):
    sample = sps.uniform.rvs(loc = -6, scale= 9, size = n)
    vsr = np.mean(sample)
    if vsr - MX > E or vsr - MX < E:
      count += 1
print(count)
```
